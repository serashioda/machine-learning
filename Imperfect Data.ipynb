# Alternate Polling Place
## DESCRIPTION
District of Columbia Alternate Polling Places. This dataset contains points representing locations of alternate polling places, created as part of the DC Geographic Information System (DC GIS) for the D.C. Office of the Chief Technology Officer (OCTO) and participating D.C. government agencies. Alternate Polling places were identified from data provided by the Board of Elections and Ethics (BOEE). These are locations if the primary polling site cannot be used. All DC GIS data is stored and exported in Maryland State Plane coordinates NAD 83 meters.
