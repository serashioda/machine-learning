```python
data = pd.read_csv(os.getcwd() + '/Cityworks_Workorders.csv')
data
```

```python
plt.title("Relationship between all initiated work orders and those completed.")
from dateutil import parser
import math

def sift(row):
    try:
        id = parser.parse(row["INITIATEDDATE"])
        cd = parser.parse(row["WORKORDERCLOSEDDATE"])
        return (cd - id).days
    except TypeError:
        return None
x = []
for v in data.apply(sift, axis=1):
    if v != None:
        x.append(v)
data = pd.DataFrame(x)
data.mean()
```

What is the percentage of the most common ticket request out of all work orders?

```python
job_count_by_type = data.DESCRIPTION.value_counts()
most_common = data.DESCRIPTION.value_counts().max()
job_count_total = data.DESCRIPTION.value_counts().sum()
most_common_percentage = most_common / job_count_total * 100
most_common_percentage
```

What is the average days to close a work order?

```python
avg_time =data.DAYSTOCLOSE.median()
print(avg_time)
# data[data.DAYSTOCLOSE == avg_time]
```
